Find the manufacturer (producer) of the plane that flew the most from JFK Airport in January, you can follow these steps:

```python
df1=(
    flights[(flights["origin"]=='JFK')&flights['month']==1]
    .groupby('tailnum', as_index=False)
    .agg(number_of_flights =('tailnum', 'count'))
    .sort_values('number_of_flights', ascending=False)
    .head(1)
 )
df1
```

```python

pd.merge(df1, planes, left_on='tailnum', right_on='tailnum')['manufacturer'][0]

```

```python
df2=(
flights[(flights["dest"]=="MIA")&(flights['month']==1)]
.groupby("tailnum", as_index=False)
.agg(av_arr_delay = ("arr_delay", "mean"))
.sort_values("av_arr_delay", ascending=False)
.head(1)
)
df2
```

Filter the flights departing from JFK Airport (origin = 'JFK') in January, calculate the average departure delay for each airline, and create a barchart  plot using Plotly and Plotnie to visualize the results.


```python
df3=(
  flights[(flights['origin']=="JFK")&(flights['month']==1)]
  .groupby('carrier', as_index=False)
  .agg(avg_dep_delay=('dep_delay', "mean"))
  .sort_values("avg_dep_delay", ascending=False)

)
df3
```
