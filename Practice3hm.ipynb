```python
#Aggregate by multiple functions:

import pandas as pd

url = 'https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/chipotle.tsv'

# Read the first few lines of the file to inspect its structure
try:
    orders = pd.read_csv(url, sep='\t', nrows=10)
    print(orders)
except pd.errors.ParserError as e:
    print(f"ParserError: {e}")

```

```python
orders.head(10)
```

```python
orders[orders.order_id == 1].item_price.sum()
```

```python
orders.groupby('order_id').item_price.sum().head()
```

```python
orders.groupby('order_id').item_price.agg(['sum', 'count']).head()
```

```python
#Check for alternative equality:
df = pd.DataFrame({'c':[1,2,3], 'd':[1.0, 2.0, 3.0], 'e': [1.0,2.0,3.000005]})
df
```
