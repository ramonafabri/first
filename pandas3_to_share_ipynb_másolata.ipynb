Exercise 1: Average Departure Delay by Carrier
Calculate the average departure delay and the total number of flights for each carrier.



```python
flights.head()
```

```python
flights["carrier"].unique()
```

```python
flights.groupby("carrier")["dep_delay"].mean()
```

```python
task5 = flights.groupby("carrier").agg({'dep_delay' : "mean", 'flight': "count"}).reset_index()
```

```python
task5.rename ({'dep_delay': "avg_dep_delay", 'flight' : "total_number_of_flights"}, axis="columns")
```
