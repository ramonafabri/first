```python
import plotly.express as px
```

```python
df = px.data.gapminder()
```

```python
df.head()
```

```python
df_us = df[df['country']=='United States']
df_us
```

```python
fig = px.bar(df_us, x='year',y='pop')
```

```python
fig.show()
```

```python
fig = px.bar(df_us, x='year',y='pop', width=950, height=500)
```

```python
fig.show()
```

```python
fig = px.bar(df_us, x='year',y='pop')
fig.update_layout(title='US population over years'
                  width=950, height=500, template='simple_white', xaxis_title='Year', yaxis_title='Population')
fig.show()
```

```python
fig = px.bar(df_us, x='year',y='pop', title='US population over years', width=950, height=500)
fig.update_layout(template='simple_white', xaxis_title='Year', yaxis_title='Population')
fig.update_traces(textposition='outside', texttemplate='%{value:.4g}')
fig.update_xaxes(type='category')
fig.add_hline()
fig.show()
```

```python
import plotly.express as px
long_df = px.data.medals_long()
long_df
```

```python
fig = px.bar(long_df, x='nation', y='count', color="medal")
fig.update_layout(title='Olympic medals', xaxis_title= 'Nation', yaxis_title='Medal count', plot_bgcolor='white', legend_title="Medal color")
fig.show()
```
